```python
import cv2
import numpy as np
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from PIL import ImageEnhance
from PIL import Image
from tensorflow import keras
from matplotlib import pyplot as plt
from IPython.display import clear_output
from keras import losses, optimizers
from keras.layers import Dense, GlobalMaxPooling2D, GlobalAveragePooling2D, Flatten, Concatenate, Dropout
from keras.models import Model
import tensorflow as tf
import re
import matplotlib.pyplot as plt
import cv2
from sklearn.model_selection import train_test_split

DATASET = re.search('.*\d+\.\d+\.\d+_.*(\d{2}x)', DATA_PATH).groups(0)


image_data = {}

INPUT_SHAPE = (224, 224, 3)
GRID_SIZE = 2
GROUP = 'Green' if re.search('Green', folder) else 'Blue'

def enhance_image(image, fl=False, GRID_SIZE = 2):
    if fl:
      GRID_SIZE = 1
    if GROUP == 'Green':
        image = image.resize((INPUT_SHAPE[0] * GRID_SIZE, INPUT_SHAPE[1] * GRID_SIZE))
        image = ImageEnhance.Brightness(image=image).enhance(6)
        image = ImageEnhance.Contrast(image=image).enhance(4)
        image = ImageEnhance.Sharpness(image=image).enhance(3)
    else:
        image = image.resize((INPUT_SHAPE[0] * GRID_SIZE, INPUT_SHAPE[1] * GRID_SIZE))
        image = ImageEnhance.Brightness(image=image).enhance(2)
        image = ImageEnhance.Contrast(image=image).enhance(2)
        image = ImageEnhance.Sharpness(image=image).enhance(2)
    return image

def load_images(df, fl=False):
    M = INPUT_SHAPE[0]
    N = INPUT_SHAPE[1]
    if not fl:
      dataset = np.ndarray(shape=(df.shape[0] * GRID_SIZE * GRID_SIZE * 2, INPUT_SHAPE[0], INPUT_SHAPE[1], 3),
                      dtype=np.float32)
      labels = np.ndarray(shape=(df.shape[0] * GRID_SIZE * GRID_SIZE * 2), dtype=np.int64)
    else:
        dataset = np.ndarray(shape=(df.shape[0], INPUT_SHAPE[0], INPUT_SHAPE[1], 3),
                      dtype=np.float32)
        labels = np.ndarray(shape=(df.shape[0]), dtype=np.int64)
    print(dataset.shape,  labels.shape)
    data_index = 0
    for index, row in df.iterrows():
        image = load_img(row['filename'])
        image = enhance_image(image, fl)
        images = []
        if not fl:
          # blurred = cv2.blur(img_to_array(image), (2, 2))
          # rotated1 = image.rotate(90)
          # rotated2 = image.rotate(np.random.randint(1, 180, 1))
          # rotated3 = image.rotate(np.random.randint(1, 270, 1))
          images.extend([image])
        else:
          images.extend([image])
        for ind, image in enumerate(images):
          img_array = img_to_array(image)
          if np.mean(img_array) <= 0.5:
            break
          tiles = [img_array[x:x+M,y:y+N] for x in range(0,img_array.shape[0],M) for y in range(0,img_array.shape[1],N)]
          for i in range(0, len(tiles)):
            dataset[data_index] = tiles[i] / 255.0
            labels[data_index] = int(row['well_number']) - 1
            image_data[data_index] = '{well_type} - {well_number} - {field}'.format(well_type=row['well_type'],
                                                                                    well_number=row['well_number'],
                                                                                    field=row['field'])
            data_index += 1
        if not fl:
          dataset[data_index] = img_to_array(image.resize((INPUT_SHAPE[0], INPUT_SHAPE[1]))) / 255.0
          labels[data_index] = int(row['well_number']) - 1
          image_data[data_index] = '{well_type} - {well_number} - {field}'.format(well_type=row['well_type'],
                                                                        well_number=row['well_number'],
                                                                        field=row['field'])
          data_index += 1
    return dataset[:data_index], labels[:data_index]

  
class PlotLearning(keras.callbacks.Callback):
    """
    Callback to plot the learning curves of the model during training.
    """
    def on_train_begin(self, logs={}):
        self.metrics = {}
        for metric in logs:
            self.metrics[metric] = []
            

    def on_epoch_end(self, epoch, logs={}):
        # Storing metrics
        for metric in logs:
            if metric in self.metrics:
                self.metrics[metric].append(logs.get(metric))
            else:
                self.metrics[metric] = [logs.get(metric)]
        
        # Plotting
        metrics = [x for x in logs if 'val' not in x]
        
        f, axs = plt.subplots(1, len(metrics), figsize=(15,5))
        clear_output(wait=True)

        for i, metric in enumerate(metrics):
            axs[i].plot(range(0, epoch + 1), 
                        self.metrics[metric], 
                        label=metric)
            if logs['val_' + metric]:
                axs[i].plot(range(0, epoch + 1), 
                            self.metrics['val_' + metric], 
                            label='val_' + metric)
                
            axs[i].legend()
            axs[i].grid()

        plt.tight_layout()
        plt.show()
```

```python
class Net:
    def __init__(self, loss, optmizer, metrics):
        self.loss = loss
        self.optimizer = optmizer
        self.metrics = metrics

    def create_model(self, output_space: int) -> Model:
        base_model = tf.keras.applications.DenseNet169(input_shape=INPUT_SHAPE,
                                               include_top=False,
                                               weights='imagenet')
        base_model.trainable = False

        out0 = base_model.output
        out = GlobalAveragePooling2D()(out0)
        
        predictions = Dense(output_space, activation="sigmoid")(out)

        model = Model(inputs=base_model.input, outputs=predictions)

        for layer in base_model.layers[:-10]:
            layer.trainable = True

        model.compile(loss=self.loss, optimizer=self.optimizer, metrics=self.metrics)

        return model

import keras.backend as K


def f1_metric(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    recall = true_positives / (possible_positives + K.epsilon())
    f1_val = 2*(precision*recall)/(precision+recall+K.epsilon())
    return f1_val
```

```python
def parse_filename(s, PATTERN):
    result = re.match(PATTERN, s)
    return result.groups() + (s,)


def display_samples(df, columns=5, rows=4):
    fig=plt.figure(figsize=(5*columns, 4*rows))

    for i in range(columns*rows):
        img = cv2.imread(df.loc[i, 'filename'])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        fig.add_subplot(rows, columns, i+1)
        plt.title('{well_type} - {well_number} - {field}'.format(well_type=df.loc[i, 'well_type'],
                                                                 well_number=df.loc[i, 'well_number'],
                                                                 field=df.loc[i, 'field']))
        plt.imshow(img)
    
    plt.tight_layout()

def display_processed_images(dataset, columns=5, rows=4):
    fig=plt.figure(figsize=(5*columns, 4*rows))

    for i in range(columns*rows):
        img = dataset[i]        
        fig.add_subplot(rows, columns, i+1)
        plt.title(image_data[i])
        plt.imshow(img)
    
    plt.tight_layout()
```
